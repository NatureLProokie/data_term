```python
import pandas as pd
import pyscipopt as opt
```

```python
PRODUCTS = ['A1','A2','B1','B2']

# 产线
LINES = ['L1', 'L2']

# 上一次生产中最后的产品
LAST_PRODUCTION = {
    'L1':'A1',
    'L2':'A2',
}

# 产品需求
DEMAND = {
    'A1':14,
    'A2':10,
    'B1':12,
    'B2':12,
}

# 切换成本
CHANGEOVER_COST = {
    ('A1', 'A1'):0,
    ('A1', 'A2'):1,
    ('A1', 'B1'):4,
    ('A1', 'B2'):4,
    ('A2', 'A1'):1,
    ('A2', 'A2'):0,
    ('A2', 'B1'):4,
    ('A2', 'B2'):4,
    ('B1', 'A1'):4,
    ('B1', 'A2'):4,
    ('B1', 'B1'):0,
    ('B1', 'B2'):1,
    ('B2', 'A1'):4,
    ('B2', 'A2'):4,
    ('B2', 'B1'):1,
    ('B2', 'B2'):0
}

# 产线产能
LINE_CAPACITY = {'L1': 24,
                 'L2': 24,}
```

```python
SLOTS = range(len(PRODUCTS))
print(SLOTS)
```

函数 add_variable_by_idx_set 用于根据索引集合 idx_set 批量创建优化模型中的变量

```python
def add_variable_by_idx_set(idx_set=None, var_name=None, var_type=None, lb=0, ub=None):
    variable = {}
    for idx in idx_set:
        variable[idx] = model.addVar(vtype=var_type, name=var_name + '_'.join(tuple(map(str,idx))), lb=lb, ub=ub)
    return variable
```

```python
def process_data(SLOTS, LINES, PRODUCTS):

    list_tmp = []
    for i in SLOTS:
        list_tmp.append(i)

    products = pd.DataFrame(PRODUCTS).rename(columns={0: 'products'})
    slots = pd.DataFrame(list_tmp).rename(columns={0: 'slots'})
    lines = pd.DataFrame(LINES).rename(columns={0: 'lines'})

    products['cols'] = 1
    slots['cols'] = 1
    lines['cols'] = 1

    line_slots = pd.merge(lines, slots, on=['cols'], how='outer')
    line_products = pd.merge(lines, products, on=['cols'], how='outer')
    line_slots_products= pd.merge(line_slots, products, on=['cols'], how='outer')
    line_slots_products_products = pd.merge(line_slots_products, products, on=['cols'], how='outer', suffixes=['_star', '_end'])

    dict_line_slots = line_slots.set_index(['lines', 'slots'])['cols'].to_dict()
    dict_line_products = line_products.set_index(['lines', 'products'])['cols'].to_dict()
    dict_line_slots_products = line_slots_products.set_index(['lines', 'slots', 'products'])['cols'].to_dict()
    dict_line_slots_products_products = line_slots_products_products.set_index(['lines', 'slots', 'products_star', 'products_end'])['cols'].to_dict()

    return dict_line_slots, dict_line_products, dict_line_slots_products, dict_line_slots_products_products, line_slots_products
```

为什么需要 isBusy = sign(quantity)？
在优化模型中，某些决策变量需要基于其他变量的状态来决定其值。例如，你可能希望：

当生产量大于 0 时：设置 isBusy 为 1，表示产线在该时段忙碌。
当生产量等于 0 时：设置 isBusy 为 0，表示产线在该时段闲置。
直接使用非线性函数（如 sign()）会使模型变得复杂或不可解。因此，通过引入 isBusy 变量和使用 Big-M 方法，可以线性化这一逻辑。

什么是 Big-M 方法？
Big-M 方法是一种用于将逻辑约束（如二进制变量与连续或整数变量的关系）转化为线性约束的技术。它使用一个足够大的常数 M 来确保逻辑关系的正确性。



quantity[idx] >= isBusy[idx]：

当 isBusy[idx] = 1 时，quantity[idx] 必须至少为 1。
当 isBusy[idx] = 0 时，quantity[idx] 只需要满足 quantity[idx] >= 0，这对非负整数变量 quantity 没有实际限制。
quantity[idx] <= M * isBusy[idx]：

当 isBusy[idx] = 1 时，quantity[idx] 需要满足 quantity[idx] <= M，这通常不会对模型产生限制，因为 M 足够大。
当 isBusy[idx] = 0 时，quantity[idx] 必须等于 0，因为 quantity[idx] <= 0。
如何选择 M 的值？
选择 M 的值时需要确保：

足够大：M 必须大于或等于模型中 quantity[idx] 的任何可能值。
尽可能小：过大的 M 可能导致数值不稳定或降低优化模型的性能。因此，选择一个尽可能小且满足条件的 M 是最佳实践。
在你的模型中，M = 100000 是一个常见的选择，因为它远大于任何实际的生产量。

```python
# 2. 不能超过产能
print("\n=== 约束 2: 不能超过产能 ===")
for line, capacity in LINE_CAPACITY.items():
    terms = [f"qty_{line}_{slot}_{product}" for slot in SLOTS for product in PRODUCTS]
    constraint_expression = " + ".join(terms)
    print(f"Constraint for line {line}: {constraint_expression} <= {capacity}")
    model.addCons(opt.quicksum(quantity[line, slot, product] for slot in SLOTS for product in PRODUCTS) <= capacity)

# 3. isBusy = sign(quantity)
print("\n=== 约束 3: isBusy = sign(quantity) ===")
for idx in dict_line_slots_products.keys():
    line, slot, product = idx
    qty_var = quantity[idx]
    isBusy_var = isBusy[idx]
    print(f"Constraint: {qty_var.name} >= {isBusy_var.name}")
    model.addCons(qty_var >= isBusy_var)
    print(f"Constraint: {qty_var.name} <= {M} * {isBusy_var.name}")
    model.addCons(qty_var <= M * isBusy_var)

# 4. 每个产线的每个时段只能允许最多一个产品
print("\n=== 约束 4: 每个产线的每个时段只能允许最多一个产品 ===")
for line_slot_idx in dict_line_slots.keys():
    line, slot = line_slot_idx
    terms = [f"isBusy_{line}_{slot}_{product}" for product in PRODUCTS]
    constraint_expression = " + ".join(terms)
    print(f"Constraint for line {line}, slot {slot}: {constraint_expression} <= 1")
    model.addCons(opt.quicksum(isBusy[line, slot, product] for product in PRODUCTS) <= 1)

# 5. 每个产线上每个产品只能出现在最多一个时段里
print("\n=== 约束 5: 每个产线上每个产品只能出现在最多一个时段里 ===")
for line_product_idx in dict_line_products:
    line, product = line_product_idx
    terms = [f"isBusy_{line}_{slot}_{product}" for slot in SLOTS]
    constraint_expression = " + ".join(terms)
    print(f"Constraint for line {line}, product {product}: {constraint_expression} <= 1")
    model.addCons(opt.quicksum(isBusy[line, slot, product] for slot in SLOTS) <= 1)

# 6. 统计每个时段被占用情况，不允许出现前面时段没有生产，后面时段有生产的情况
print("\n=== 约束 6: 统计每个时段被占用情况 ===")

# 第一部分：确保后续时段被占用，前面的时段也被占用
print("\n--- Constraint: No production in later slots without production in earlier slots ---")
for line, slot in dict_line_slots:
    if slot != 0:
        print(f"Constraint for line {line}, slot {slot}: slotBusy_{line}_{slot-1} >= slotBusy_{line}_{slot}")
        model.addCons(slotBusy[line, slot-1] >= slotBusy[line, slot])

# 第二部分：定义 slotBusy_sum 和关联 slotBusy
print("\n--- Constraint: Define slotBusy_sum and relate it to slotBusy ---")
for line, slot in dict_line_slots:
    # Define slotBusy_sum
    print(f"Constraint for slotBusy_sum_{line}_{slot}: slotBusy_sum_{line}_{slot} == sum(isBusy_{line}_{slot}_{product} for product in PRODUCTS)")
    model.addCons(slotBusy_sum[line, slot] == opt.quicksum(isBusy[line, slot, product] for product in PRODUCTS))
    
    # slotBusy >= slotBusy_sum
    print(f"Constraint for slotBusy_{line}_{slot}: slotBusy_{line}_{slot} >= slotBusy_sum_{line}_{slot}")
    model.addCons(slotBusy[line, slot] >= slotBusy_sum[line, slot])
    
    # slotBusy <= M * slotBusy_sum
    print(f"Constraint for slotBusy_{line}_{slot}: slotBusy_{line}_{slot} <= {M} * slotBusy_sum_{line}_{slot}")
    model.addCons(slotBusy[line, slot] <= M * slotBusy_sum[line, slot])
```

```python
# 7. 统计每个时段的切换情况
print("\n=== 约束 7: 统计每个时段的切换情况 ===")

# SLOTS[0]
print("\n--- SLOTS[0] ---")
for line, product in dict_line_products:
    if product == LAST_PRODUCTION[line]:
        # 打印约束内容
        constraint_str = f"changeOver_{line}_0_{LAST_PRODUCTION[line]}_{product} == 0"
        print(f"Constraint: {constraint_str}")
        # 添加约束
        model.addCons(changeOver[line, 0, LAST_PRODUCTION[line], product] == 0)
    else:
        # 打印约束内容
        constraint_str = f"changeOver_{line}_0_{LAST_PRODUCTION[line]}_{product} == isBusy_{line}_0_{product}"
        print(f"Constraint: {constraint_str}")
        # 添加约束
        model.addCons(changeOver[line, 0, LAST_PRODUCTION[line], product] == isBusy[line, 0, product])

# SLOTS[1:]
print("\n--- SLOTS[1:] ---")
for line, slot, product_1, product_2 in dict_line_slots_products_products:
    if slot != 0:
        if product_1 == product_2:
            # 打印约束内容
            constraint_str = f"changeOver_{line}_{slot}_{product_1}_{product_2} == 0"
            print(f"Constraint: {constraint_str}")
            # 添加约束
            model.addCons(changeOver[line, slot, product_1, product_2] == 0)
        else:
            # 为每个切换组合创建唯一的辅助变量名
            var_name = f"and_isBusy_{line}_{slot}_{product_1}_{product_2}"
            print(f"Constraint: {changeOver[line, slot, product_1, product_2].name} == {var_name}")
            print(f"  Creating auxiliary variable: {var_name}")
            # 创建辅助二进制变量
            and_isBusy = model.addVar(vtype='B', name=var_name)
            
            # 定义 isBusy_tmp 表达式
            isBusy_tmp_expr = isBusy[line, slot-1, product_1] + isBusy[line, slot, product_2] - 2 + B
            print(f"  Constraint: isBusy_tmp = isBusy_{line}_{slot-1}_{product_1} + isBusy_{line}_{slot}_{product_2} - 2 + {B}")
            
            # 添加约束 isBusy_tmp <= and_isBusy * M
            constraint1_str = f"isBusy_tmp ({isBusy[line, slot-1, product_1].name} + {isBusy[line, slot, product_2].name} - 2 + {B}) <= {and_isBusy.name} * {M}"
            print(f"  Constraint: {constraint1_str}")
            model.addCons(isBusy[line, slot-1, product_1] + isBusy[line, slot, product_2] - 2 + B <= and_isBusy * M)
            
            # 添加约束 and_isBusy * M <= M + isBusy_tmp - B
            constraint2_str = f"{and_isBusy.name} * {M} <= {M} + (isBusy_tmp) - {B}"
            print(f"  Constraint: {constraint2_str}")
            model.addCons(and_isBusy * M <= M + isBusy[line, slot-1, product_1] + isBusy[line, slot, product_2] - 2 + B - B)
            
            # 关联 changeOver 与 and_isBusy
            constraint3_str = f"changeOver_{line}_{slot}_{product_1}_{product_2} == {and_isBusy.name}"
            print(f"  Constraint: {constraint3_str}")
            model.addCons(changeOver[line, slot, product_1, product_2] == and_isBusy)
```

```python
model.setObjective(opt.quicksum(changeOver[line, slot, product_1, product_2] * CHANGEOVER_COST[product_1, product_2]
                                for line, slot, product_1, product_2 in dict_line_slots_products_products))
model.setMinimize()
model.optimize()
```
